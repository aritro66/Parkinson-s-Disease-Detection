Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm;
```

Collecting and Processing Data

```python
# Loading the dataset using Pandas function
disease_data = pd.read_csv('/content/parkinsons.csv')
```

```python
# Checking data of first 5 rows
disease_data.head()
```

```python
# Checking data of last 5 rows
disease_data.tail()
```

```python
# No. of rows and columns
disease_data.shape
```

```python
# Statistical info
disease_data.describe()
```

```python
# Getting basic information about data
disease_data.info()
```

```python
# Finding missing values
disease_data.isnull().sum()
```

```python
# number of values for status
disease_data["status"].value_counts()
```

1 represents suffering from Parkinson's disease

0 represents not suffering from Parkinson's disease

```python
# grouping the data based on the target variable
disease_data.groupby('status').mean()
```

```python
# Separating data and labels
X = disease_data.drop(columns=['status','name'],axis=1)
Y = disease_data['status']
print(X.shape)
print(Y.shape)
```

Training and Testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Data Standardization

```python
scalar = StandardScaler()
```

```python
scalar.fit(X_train)
```

```python
# Transforming Data
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

Modal Training using SVM

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,y_train)
```

Modal Evalution

```python
# Accuracy of Training Data
train_prediction = classifier.predict(X_train)
print('Accuracy of Training Data : ',accuracy_score(y_train,train_prediction))
```

```python
# Accuracy of Testing Data
test_prediction = classifier.predict(X_test)
print('Accuracy of Testing Data : ',accuracy_score(y_test, test_prediction))
```

Making a Prediction using certain input

```python
input_data=np.array([119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654])

# Reshaping Array
input_data=input_data.reshape(1,-1)
# Standardize Array
input_data=scalar.transform(input_data)
prediction = classifier.predict(input_data)

if(prediction[0]==0):
  print("Not suffering from Parkinson's disease.")
else:
  print("suffering from Parkinson's disease.")
```
